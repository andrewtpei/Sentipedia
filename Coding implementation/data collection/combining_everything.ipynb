```python
import pandas as pd

df_content = pd.read_csv('df_news_content.csv')
df_summary = pd.read_csv('summary.csv')
df_reddit_news = pd.read_csv('merged.csv')
```

```python

merged_df = pd.merge(df_content, df_summary, left_on='text', right_on='Original Paragraph', how='inner')


merged_df.drop('Original Paragraph', axis=1, inplace=True)
merged_df.drop('Unnamed: 0', axis=1, inplace=True)

merged_df
```

```python

new_df = df_reddit_news.merge(merged_df, on=['title', 'text'], how='inner')
new_df.to_csv("final.csv", index=False)
```
